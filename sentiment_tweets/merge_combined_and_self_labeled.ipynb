```python
import csv
import pandas as pd
```

```python
merged_combined = pd.read_csv("mycsvfileCombined.csv")
randomly_sampled = pd.read_csv("randomly_sample_01_train.csv")
```

```python
merged_combined
```

```python
randomly_sampled = randomly_sampled[['tweet_id', 'text', 'sentiment']]
randomly_sampled.rename(columns={'text':'text_clean'}, inplace=True)
randomly_sampled.rename(columns={'sentiment':'label'}, inplace=True)
```

```python
# Field 'attitude': 0 denotes positive attitude, 1 denotes neutral attitude, 2 denotes negative attitude
randomly_sampled['label'] = randomly_sampled['label'].replace({'POSITIVE':'0', 'Positive':'0'})
randomly_sampled['label'] = randomly_sampled['label'].replace({'NEUTRAL':'1', 'Neutral':'1'})
randomly_sampled['label'] = randomly_sampled['label'].replace({'NEGATIVE':'2', 'Negative':'2'})
randomly_sampled
```

```python
# swapping columns
randomly_sampled = randomly_sampled[['tweet_id','label','text_clean']]
```

```python
randomly_sampled
```

```python
new_merged = merged_combined.append(randomly_sampled, ignore_index=True)
```

```python
new_merged.to_csv("new_merged.csv", sep=',')
```
