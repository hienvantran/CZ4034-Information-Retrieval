```python
import numpy as np
import csv
import pandas as pd
import os
import re
import string as st
```

```python
# Function to remove @names and special character
def punc_clean(text):
    
    text= ' '.join(re.sub("(@[A-Za-z0-9_]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)" , " ", text).split())
    
    return text

def preprocess(path):
    path = os.path.join(path)
    data = pd.read_csv(path)
    
    data['text'] = data['text'].apply(punc_clean)
    data['creation_date_time'] = data['creation_date_time'].apply(lambda x: x[:10])
    data["location"].fillna("No Location", inplace = True)
    
    return data
```

```python
pfizer_data = preprocess("latest_data/Pfizer.csv")
pfizer_data
```

```python
# Convert it to CSV
pfizer_data.to_csv("pfizer.csv", encoding='utf-8', index=False)
```
